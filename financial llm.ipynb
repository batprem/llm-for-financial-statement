```python
from IPython.display import Markdown

def chat(text: str) -> str:
    result = ""
    response = ollama.chat(
        model='gemma2:latest', stream=True,
        messages=[{'role': 'user', 'content': text}]
    )

    for r in response:
        result += r["message"]["content"]
        print(r["message"]["content"], end="")
    return result


answer = chat(f"""The following text is finanical statement.

{df_prompt}

Act as a financial expert and give the user your opinion about your thought toward to this company by break down positive and negetive sign.
""")

Markdown(answer)
```
